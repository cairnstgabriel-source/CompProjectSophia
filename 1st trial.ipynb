```python
import numpy as np
import matplotlib.pyplot as plt
# putting particles in 2D lattice with 4 particles
L = 4
def simple_lattice_2D(L,n):
     positions = []
     N = int(np.sqrt(n))
     separation = L / N
     for i in range(0,N):
            for j in range(0,N):
                x = (i + 0.5) * separation
                y = (j + 0.5) * separation
                positions.append((x,y))
     return positions


def simple_lattice_3D(L,n):
    positions = []
    N = int(np.cbrt(n))
    separation = L / N
    for i in range(0,N):
        for j in range(0,N):
            for k in range(0,N):
                x = (i + 0.5) * separation
                y = (j + 0.5) * separation
                z = (k + 0.5) * separation
                positions.append((x,y,z))
    return positions

vol_fraction = 0.72
def particle_diameter(L, n, vol_fraction):
    volume_per_particle = (L**3 * vol_fraction) / n
    diameter = 2*(3 * volume_per_particle / (4 * np.pi))**(1/3)
    return diameter
```
