```python
import numpy as np
import matplotlib.pyplot as plt

# defining simple lattice functions to place n particles in a box of length L 

def simple_lattice_2D(L,n):
     positions = []
     N = int(np.sqrt(n))
     separation = L / N
     for i in range(0,N):
            for j in range(0,N):
                x = (i + 0.5) * separation
                y = (j + 0.5) * separation
                positions.append((x,y))
     return positions


def simple_lattice_3D(L,n):
    positions = []
    N = int(np.cbrt(n))
    separation = L / N
    for i in range(0,N):
        for j in range(0,N):
            for k in range(0,N):
                x = (i + 0.5) * separation
                y = (j + 0.5) * separation
                z = (k + 0.5) * separation
                positions.append((x,y,z))
    return positions

def particle_diameter(L, n, vol_fraction):
    volume_per_particle = (L**3 * vol_fraction) / n
    diameter = 2*(3 * volume_per_particle / (4 * np.pi))**(1/3)
    return diameter

x_coords, y_coords = zip(*simple_lattice_2D(L,100))
plt.figure()
plt.plot(x_coords, y_coords, 'o')
plt.savefig('simple_lattice_2D')
```

```python
# create stepwise function to move random particle in random direction by displacement amount specified
#also account for the periodic boundary conditions inside the movement 
import random

def move_particle(positions, displacement, L):
    particle = random.randint(0, len(positions)-1)
    x_or_y = random.randint(0,1)
    pos_or_neg = random.randint(0,1)
    if x_or_y == 0:
        if pos_or_neg == 0:
            positions[particle][0] += displacement
        else:
            positions[particle][0] -= displacement
    else:
        if pos_or_neg == 0:
            positions[particle][1] += displacement
        else:
            positions[particle][1] -= displacement
    #series of if statements to account for periodic boundary conditions
    if  positions[particle][0] > L:
        positions[particle][0] -= L
    if positions[particle][0] < 0:
        positions[particle][0] += L
    if  positions[particle][1] > L:
        positions[particle][1] -= L
    if positions[particle][1] < 0:
        positions[particle][1] += L
        
    return positions

    
```
