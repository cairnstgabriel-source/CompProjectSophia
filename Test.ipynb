```python

import numpy as np
import pandas as pd
from scipy import stats


print("All libraries imported successfully!")
```

```python
np.random.seed(42)
group_A = np.random.normal(loc=5, scale=2, size=50)
group_B = np.random.normal(loc=6, scale=2.5, size=50)

df = pd.DataFrame({"Group A": group_A, "Group B": group_B})
df.head()
```

```python
t_stat, p_value = stats.ttest_ind(group_A, group_B)
print(f"T-test result: t = {t_stat:.2f}, p = {p_value:.4f}")
```

```python
df.loc['Mean'] = df.mean()
df.loc['Std'] = df.std()
df
```
